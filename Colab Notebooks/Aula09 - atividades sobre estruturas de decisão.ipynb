#Atividade com aplicação de testes condicionais.

Crie um programa em Python que a partir do peso e altura de uma pessoa, calcule e informe o IMC arredondado, informando tambem o estado.

##Exercicio 3

Refaça o programa usando a função Match Case, que está disponível no Python a partir da
versão 3.10

```python
peso = float(input('Digite seu peso: '))
altura = float(input('Digite sua altura: '))

imc = peso / (altura * altura)

print(f'\nIMC: {round(imc,1)}\n')

match imc:
    case imc if imc < 17:
         print('Muito abaixo do peso')
    case imc if imc >= 17 and imc <= 18.49:
         print('Abaixo do peso')
    case imc if imc >= 18.49 and imc <= 24.99 :
         print('Peso normal')
    case imc if imc >= 25 and imc <= 29.99:
         print('Acima do peso')
    case imc if imc >= 30 and imc <= 34.99:
         print('Obesidade I')
    case imc if imc >= 35 and imc <= 39.99:
         print('Obesidade II')
    case imc if imc > 40:
         print('Obesidade III')
```

#Exercicio 2

Usando as funções if - else.

```python
peso = float(input('Digite seu peso: '))
altura = float(input('Digite sua altura: '))

imc = peso / (altura * altura)

print(f'\nIMC: {round(imc,1)}\n')

if imc < 17:
  print('Muito abaixo do peso')
else:
  if imc >= 17 and imc <= 18.49:
    print('Abaixo do peso')
  else:
    if imc >= 18.49 and imc <= 24.99 :
      print('Peso normal')
    else:
      if imc >= 25 and imc <= 29.99:
          print('Acima do peso')
      else:
         if imc >= 30 and imc <= 34.99:
             print('Obesidade I')
         else:
            if imc >= 35 and imc <= 39.99:
              print('Obesidade II')
            else:
              if imc > 40:
               print('Obesidade III')
```
