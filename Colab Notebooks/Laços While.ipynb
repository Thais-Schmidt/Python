# **While**

### *While* significa *Enquanto* em português

### A sintaxe para esta função do python é:

## **while** condição **:**
*   Código
*   Código
*   Código


#### Atentar para a indentação dos códigos dentro do while. Vejamos um exemplo para ilustrar melhor:

```python
# While repete um bloco de código, enquanto a condição for verdadeira

x= 1

while x < 5:
  print(x)
  x = x + 1

# Aqui já acabou o loop while

print("O valor de x após o final do loop, depois que saiu, vale = ", x)
```

# Instruções ***pass***, ***break*** e ***continue***

### Essas instruções são uteis para se trabalhar com exceções

 ***pass*** - é utilizado quando não queremos executar nenhum código,
 simplesmente não vai fazer nada.

 ***break*** - a instrução break é colocada dentro de um bloco de  loop para forçar a saída desse loop. Normalmente, usamos após uma instrução condicional if.

 ***continue*** - a instrução continue dá a opção de ignorar a parte final de um loop, usando uma segunda condição. Ou seja, a iteração atual do loop será interrompida, mas o programa retornará ao início do loop e completar o resto desse loop..

 Em resumo:
*   ***pass*** = passa reto, continuando o código.
*   ***break*** = quebra o loop, sai fora.
*   ***continue*** = pula a vez, continua para a próxima iteração.

```python
# Exemplo de uso do pass dentro do loop while

# Mostrar os números de 0 a 9, mas se encontrar o número 4, não fazer nada.

valor = 0

while valor < 10:
  if valor == 4:
    pass
  else:
    print("dentro do if = ", valor)
  valor = valor + 1
  print("Valor incrementado = ", valor)
```

```python
# Exemplo de uso do break dentro do loop while

# Mostrar os números de 0 a 9, mas se encontrar o número 4, interrompa o loop.
# Observe que o break cortará o loop, não mpstrando o 4

valor = 0

while valor < 10:
  if valor == 4:
    break
  print(valor)
  valor = valor + 1
  print("Valor incrementado = ", valor)
```

```python
# Exemplo de uso do continue dentro do loop while
# Mostrar os números de 1 a 10 e calular o quadrado e o cubo desse número, porém,
# se encontrar o número 5, não fazer nada, pular para o próximo número.

valor = 0

while valor < 10:
  valor = valor + 1
  if valor == 5:
    continue
  print("Valor incrementado = ", valor)
  print("O quadrado do valor é: ", valor**2)
  print("O cubo do valor é: ", valor**3)
  print()
```

## Também é possivel usar a cláusula **ELSE** após encerrar o loop While


```python
x = 0

while x < 10:
   print('O valor de x nesta iteração é:', x)
   print(' X ainda é menor que 10, somando 1 a x')
   x += 1  # é equivalente a x = x + 1

else:
     print('Loop concluído')
print("O valor de X agora é",x)
```

## A cláusula ***else*** não é executada quando o loop é quebrado pelo break.

```python
# Podemos utilizar o break para interromper o loop.
# neste exemplo, se encontrarmos o número 4, interrompemos o loop.
# observe que o "Loop concluído" não será exibido pelo else pois o loop foi
# quebrado pelo break.

# Mude o 4 para 3.5 e verifique o que acontecerá.

valor = 0

while valor < 10:
  if valor == 4:
    break  # encerra o loop de valor = 4
  print(valor)
  valor = valor + 1
else:
  print('Loop concluído')
```

### **Exercício 4:** Fazer um programa em Python utilizando a estrutura while que leia 10 valores inteiros e:

*   Encontre e mostre o maior valor
*   Encontre e mostre o menor valor
*   Calcule e mostre a média dos números lidos

```python
# Declaração de variáveis
maior_valor = None
menor_valor = None
soma = 0
contador = 0

# Laço while para ler 10 valores
while contador < 10:
  # Entrada de um valor inteiro
  valor = int(input("Digite um valor inteiro: "))

  # Atualização do maior valor
  if maior_valor is None or valor > maior_valor:
    maior_valor = valor

  # Atualização do menor valor
  if menor_valor is None or valor < menor_valor:
    menor_valor = valor

  # Acumulação da soma para calcular a média
  soma += valor

  # Incremento do contador
  contador += 1

# Cálculo da média
media = soma / contador

# Apresentação dos resultados
print(f"O maior valor é: {maior_valor}")
print(f"O menor valor é: {menor_valor}")
print(f"A média dos valores é: {media}")
```

### **Exercício 5:** Fazer um programa em linguagem Python utilizando while que pergunte a temperatura dos pacientes e imprima uma mensagem de:

*   Estado normal (temperatura menor que 37,2 C)
*   Estado febril (37,3 C a 38 C)
*   Estado com febre (38 C a 39 C)
*   Estado com febre alta(acima 39 C)

Após mostrar a mensagem, perguntar se o usuário deseja continuar com mais um paciente.
No final, mostre a quantidade de pessoas analisadas e a média das temperaturas delas.

```python
# Variáveis para contagem e soma de temperaturas
contador_pessoas = 0
soma_temperatura = 0
continuar = "s" # variável de controle, inicializada para entrar no loop

# Loop while para receber a temperatura de cada cliente
while continuar == "s":

    # Solicita a temperatura do cliente
    temperatura_cliente = float(input("Digite a temperatura do cliente (em Celsius): "))

    # Verifica o estado febril e imprime a mensagem correspondente
    if temperatura_cliente < 37.2:
        print("Temperatura normal.")
    elif temperatura_cliente <= 38:
        print("Estado febril.")
    elif temperatura_cliente <= 39:
        print("Febre.")
    else:
        print("Febre alta!")

    # Atualiza as variáveis de contagem e soma
    contador_pessoas += 1
    soma_temperatura += temperatura_cliente

    # Pergunta se o usuário deseja continuar
    continuar = input("Deseja analisar a temperatura de outro cliente? (s/n): ").lower()

# Calcula a média de temperatura
media_temperatura = soma_temperatura / contador_pessoas

# Imprime a quantidade de pessoas analisadas e a média de temperatura
print(f"\nTotal de pessoas analisadas: {contador_pessoas}")
print(f"Média de temperatura: {media_temperatura:.2f}°C")
```
