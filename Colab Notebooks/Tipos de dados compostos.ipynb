# Tupla
### É uma coleção de elementos cujo objetivo é ser um conjunto de **dados fixos** agrupados, **somente para consulta**.
### Pode ser delimitado por parenteses (opcional) e pode conter objetos de *tipos diferentes* em uma mesma tupla.
### *Permite somente a leitura dos dados da tupla, não permite a alteração desses dados.*

```python
tupla1 = ('hello','bye','Tchüss','ciao','Hallo','hi','hello')
print(tupla1)
print(type(tupla1))
```

```python
tupla2 = 'hello','bye','Tchüss','ciao','Hallo','hi','hello'
print(tupla2)
print(type(tupla2))
```

```python
# Observe que as duas tuplas são realmente iguais
tupla1 == tupla2
```

```python
# Método para contagem de valores na tupla
n = tupla1.count('hello')
print("Nº de hello :", n)
```

```python
# Método para achar o índice do primeiro valor encontrado na tupla
tupla1.index('bye')
```

## Criando uma tupla com elementos de vários tipos.

```python
tupla3 = (1, 3.1416, +123.E-2, "qualquer coisa", (1, 2, 3), 5+3j)

# Lendo a tupla e verificando seu tipo
print(tupla3)
print(type(tupla3), "\n")
```

## Imprimindo os elementos da tupla (leitura)

#### Em Python os índices sempre começam por 0 com o primeiro elemento.

*   Portanto o elemento de índice N será o elemento N+1 da sequência de dados
*   Se você quiser o enésimo elemento da sequência deverá chamar o índice N-1

#### Vejamos os exemplos para ficar mais claro.

```python
# Lendo o primeiro elemento da tupla, que tem índice [0]
print(tupla3[0])
print("O tipo do primeiro elemento é ", type(tupla3[0]), "\n")

# Lendo o segundo elemento da tupla, que tem índice [1]
print(tupla3[1])
print("O tipo do segundo elemento é ", type(tupla3[1]), "\n")
```

```python
# N = tamanho da tupla3 = quantidade de elementos retornado pela função len()
N = len(tupla3)
print("O N da tupla é", N)
print("O último elemento da tupla é", tupla3[N-1])
```

## Criando uma tupla vazia ou com um único elemento
### Tuplas vazias são construídas por um par de parênteses vazios.
### Uma tupla unitária é construída por um único valor e uma vírgula entre parênteses (não basta colocar um único valor entre parênteses)

```python
tupla_vazia = ()
print(type(tupla_vazia))
print(len(tupla_vazia))
print(tupla_vazia)
```

```python
tupla_unitaria = ('sozinho aqui',)
print(type(tupla_unitaria))
print(len(tupla_unitaria))
print(tupla_unitaria)
```

# LISTA

### É uma coleção de elementos de qualquer tipo e que podem ser alterados.
### Deve ser delimitado por colchetes e pode conter objetos de *tipos diferentes* em uma mesma lista.
### Podemos inclusive fazer listas de lista, o que equivale a uma matriz.

```python
alunos = ['Alex', 'Anderson', 'André', 'Álvaro', 'Alisson', 'André']
notas = [10, 8.5, 7.8, 8.0, 8.2, 7.9]
print(alunos)
print(type(alunos), "\n")

print(notas)
print(type(notas))
```

```python
# Imprimindo o primeiro nome e sua nota
print("A nota do",alunos[0],"é", notas[0])
```

```python
# Alterando a nota do segundo aluno para 9.5
notas[1] = 9.5
print(notas)
```

```python
# Somando 1 na nota do último aluno
notas[-1] = notas[-1] + 1
print(notas)
```

```python
# Verificando a quantidade de alunos
print(len(alunos))
```

### Suponha agora que o aluno André desistiu do curso e precisa ser removido da lista.
### Lembre-se que a nota dele também precisa ser removida da lista de notas.

```python
nome = 'André'
indice_aluno = alunos.index(nome)    # Descobre o índice do aluno na lista
print("O aluno", nome, "é o aluno número", indice_aluno + 1, "da lista e sua nota é", notas[indice_aluno] )
alunos.remove(nome)   # o método list.remove('valor') remove somente o 1º elemento = valor
notas.pop(indice_aluno)  # o método list.pop(índice) remove o item de índice especificado
print("O aluno", nome, "e sua nota foram removidos da lista")
print(alunos)
print(notas)
```

## Exercício 1:
Usando as listas anteriores de alunos e de notas, fazer um programa que pergunte o nome do aluno a excluir e verifique se o aluno está na lista antes de excluir, a fim de se evitar erro na execução.
Para verificar se o aluno está na lista, use a função ***in***

```python
nome = input("Digite o nome do aluno a ser excluído: ")
if nome in alunos:
    indice_aluno = alunos.index(nome)
    alunos.remove(nome)
    notas.pop(indice_aluno)
    print("O aluno", nome, "e sua nota foram removidos da lista")
else:
    print("O aluno", nome, "não está na lista")
print(alunos)
print(notas)
```

## Podemos filtrar a lista usando os índices entre os colchetes.
### Veremos como fazer isto através dos exemplos a seguir.

```python
frutas = ['banana', 'maçã', 'abacaxi', 'uva', 'pêssego', 'jaca', 'caqui', 'acerola']
print(frutas)
```

### Mostrando os elementos em um intervalo de índices dado.

```python
# Mostra os dados da lista no intervalo de 1 (incluído) e 3 (não íncluido)
# Portanto só mostrará os elementos 1 (segundo elemento) e 2 (terceiro elemento)
print(frutas[1:3])
```

### Mostrando os elementos em um intervalo de índices dado porém com **passo** especificado.

### Para entender os exemplos abaixo, observe que em Python:
#### - O primeiro elemento da lista tem índice **zero - 0**
#### - O elemento indicado pelo primeiro índice é incluído
#### - O elemento indicado pelo último índice **NÃO** é incluído

```python
# Mostra os dados no intervalo especificado com passo igual a 2
# irá mostrar os elementos tomados de 2 em 2, começando do elemento 0
print(frutas[0:5:2])
```

```python
# Mostra os dados com passo invertido (-1)
print(frutas[5:0:-1])
```

## Para juntar duas listas com os elementos de cada posição formando um par, podemos utilizar a função **zip(lista1,lista2)**, que junta as duas listas em lista de **duas tuplas**.

```python
lista1 = ['banana', 'maçã', 'abacaxi', 'uva', 'pêssego']
lista2 = [1,2,3,4,5]
objetozip= zip(lista1,lista2)
lista3 = list(zip(objetozip))
print(lista3)
```

### No código acima usamos a função de conversão de objeto para lista **list()** porque a função **zip(lista1,lista2)** retorna um "objeto zip" e não uma lista.
### Essa função pode ser usada para converter uma tupla em lista também.

```python
print("O objeto é:")
print(objetozip)
print("\nO tipo do objeto é:")
print(type(objetozip))
```

```python
print("O tamanho da lista é:")
print(len(lista3))
```

## sua_lista.sort()
#### Método para ordenar a sua_lista.
Este método é do objeto sua_lista, portanto vai alterar o objeto sua_lista de forma que ele fique ordenado.

```python
frutas = ['jabuticaba', 'banana', 'maçã', 'abacaxi', 'uva', 'pêssego']
print(frutas)
```

```python
frutas.sort()
print(frutas)
```

```python
#key=len ordena pelo tamanho da palavra
frutas.sort(key=len)
print(frutas)
```

## Função sorted
## nova_lista = sorted(sua_lista)
#### Método para gerar uma lista ordenada a partir da sua_lista.
Esta é uma função que vai retornar um novo objeto, que será a sua_lista ordenada, portanto a variável nova_lista será um objeto novo, preservando a sua_lista intacta.

```python
# O sorted necessita da criação de outra lista
frutas2 = []
print(frutas2)
frutas2 = sorted(frutas)
print(frutas2)
```

```python
# O sorted também ordena pelo tamanho da palavra,  key=len
frutas2 = sorted(frutas, key=len)
print(frutas2)
```

# SET (CONJUNTO)

### Python também inclui um tipo de dados para conjuntos, chamado set.
### Um conjunto é uma coleção **desordenada** de elementos, sem elementos repetidos.
### Usos comuns para conjuntos incluem a *verificação eficiente da existência de objetos e a eliminação de itens duplicados*.
### *Conjuntos também suportam operações matemáticas como união, interseção, diferença e diferença simétrica.*
### Deve ser delimitado por chaves e pode conter objetos simples de *tipos diferentes* em uma mesma lista. (pode conter tuplas mas não pode conter listas e dicionários)
### #set {} - mutável, não tem index
### conjunto_vazio = set()

```python
conj_frutas = {'uva', 5+3j, 'jabuticaba', (1,2,3), 'banana', 'pêssego', 'maçã', 'abacaxi', 'uva', 'pêssego', 'maçã', 'banana'}
print(conj_frutas)                      # Observe que as frutas repetidas foram removidas do conjunto
```

```python
'maçã' in conj_frutas                 # O teste é mais rápido em conjuntos
```

```python
'manga' in conj_frutas
```

## A função set() pode ser usada para gerar um conjunto dos caracteres de uma string.

## Podemos fazer as operações de conjunto com as letras de duas strings.
### Vejamos um exemplo:

```python
conj_a = set ('abacaxi')
conj_b = set ('abacate')
print("Conjunto a =", conj_a)
print("Conjunto b =", conj_b)
```

```python
# Conjunto diferença: é formado pela exclusão dos elementos de conj_b que também pertencem ao conjunto conj_a
# Letras em conj_a mas não em conj_b
conj_a - conj_b
```

```python
# Conjunto diferença: é formado pela exclusão dos elementos de a que também pertencem ao conjunto a (Letras em a mas não em b)
conj_b - conj_a
```

```python
# conj_a união com conj_b (Letras em conj_a ou conj_b ou ambos, todas as letras de conj_a e conj_b)
conj_a | conj_b
```

```python
# conj_a intersecção com conj_b (letras que pertencem ao conjunto conj_a e ao conjunto conj_b ao mesmo tempo)
conj_a & conj_b
```

```python
# diferença simétrica entre os conj_a e conj_b
# Letras no conj_a ou no conj_b mas não em ambos
conj_a ^ conj_b
```

# DICIONÁRIO

### É uma estrutura de dados em Python apropriada para lidar com pares de dados como na tabela abaixo por exemplo.

###**Produto** ----**Preço**
###Alface ------R 5,00
###Batata ------R 6,20
###Cebola ------R 7,00
###Tomate ----R 10,00

### O produto vai ser a **chave** de identificação do dado e preço vai ser o **valor** associado à chave.

## Os ***dicionários*** são utilizados para agrupar elementos através da estrutura de **chave e valor**, onde a chave é o primeiro elemento seguido por dois pontos e pelo valor.

```python
feira = {
    'Alface': 5.00,
    'Batata': 6.20,
    'Cebola': 7.00,
    'Tomate': 10.00
}
print(feira)
print(type(feira))
```

```python
altura = {'Alex': 1.75, 'Valdir': 1.67, 'André': 1.70, 'João': 1.80}
peso = {'Alex': 88, 'André': 78, 'João': 95}

print(altura)
print(type(altura), '\n')

print(peso)
print(type(peso))
```

### A vantagem é na leitura dos valores com uma chamada fácil de entender.

```python
print(feira['Alface'])
print(feira['Batata'])
```

```python
print(altura['Alex'])
print(peso['João'])
```

```python
# Código fácil de interpretar
print("O Alex tem", altura['Alex'],"m de altura e pesa", peso['Alex'], 'kg' )
```

### Acrescentando dados no dicionário


```python
feira['Cenoura'] = 1.50
feira         # Mostrando o valor do dicionário sem o print
```

### Apagando dados no dicionário

```python
del(feira['Batata'])
print(feira)     # Mostrando o valor do dicionário com o print
```

```python
feira['Cenoura']             # Este valor não permenecerá na tela
print(feira['Alface'])
feira['Alface']
```

```python
# Lista das pessoas no dicionário ( Chamando as chaves = keys)
pessoas = list(altura.keys())
print(pessoas)
```

```python
# Lista em ordem alfabética
pessoas = sorted(altura.keys())
print(pessoas)
```
