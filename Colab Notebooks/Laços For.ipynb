##A sintaxe da função range() é:

range(inicio, parada, incremento)

```
# inicio = o primeiro numero é incluido na série
# parada = o ultimo numero não é incluido na série
# incremento = é o passo da contagem da série
```
Exemplo:

Definindo: inicio = 5 ;
           parada = 10

````python
for i in range(5,10):
  print(i, end=", ")
````

````python
#Exemplo com 3 argumentos
# range( inicio, parada, incremento )

for i in range(-5, 5, 2):
  print(i, end=", ")
````

````python
#Fazer contagem regressiva de 10 a 0

for i in range(10, -1, -1):
  print(i, end=", ")
````

````python
#Tabuada do 2 ao 5

for numero_coluna1 in range(2,11):
  print("______________")
  print("Tabuada do ", numero_coluna1)
  for numero_coluna2 in range(11):
    print(numero_coluna1, "x", numero_coluna2, "=", numero_coluna1 * numero_coluna2)
````

````python
#Loop em lista de listas para encontrar o manior numero
matriz = [[11,22,33], [44,55,66], [77,88,99]]
maior_numero = 0
#Loop externo
for linha in matriz:
  #Loop interno
  for num in linha:
    #Condicional
    if num > maior_numero:
     maior_numero = num
print("Maior numero:", maior_numero)
````
