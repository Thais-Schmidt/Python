#5º Exercicio

O preço ao consumidor de um carro novo é a soma do custo de fábrica com a porcentagem do distribuidor e dos impostos, ambos aplicados ao custo de fábrica. Supondo que a porcentagem do distribuidor seja de 12% e a dos impostos de 45%, prepare um algoritmo para ler o custo de fábrica do carro e imprimir o preço ao consumidor.

```python
custoCarro = float(input("Digite o custo de fabrica do carro novo:"))
print("O preço final que será cobrado do consumidor é: ", ((0.12 + 0.45) * custoCarro) + custoCarro )
```

#6º Exercicio

O cardápio de uma lanchonete é dado abaixo.
Prepare um algoritmo que leia a quantidade de cada item que você consumiu e calcule a conta final.

Hambúrguer................. R$ 15,00

Cheeseburger.............. R$ 18,00

Fritas............................ R$ 8,50

Refrigerante................. R$ 5,50

Milkshake..................... R$ 12,90

```python
print("Vamos calcular quanto deu sua conta:")
ham = float(input("Quantidade de hamburgueres:"))
che = float(input("Quantidade de cheeseburger:"))
fri = float(input("Quantidade de fritas:"))
ref = float(input("Quantidade de refrigerantes:"))
mil = float(input("Quantidade de milkshakes:"))

hamburguer = 15
cheeseburger = 18
fritas = 8.5
refrigerante = 5.5
milkshake = 12.90

print("O valor total da conta a ser paga é: ", (ham * hamburguer) + (che * cheeseburger) + (fri * fritas) + (ref * refrigerante) + (mil * milkshake))
```

#7º Exercicio

Uma companhia de carros paga a seus empregados um salário de R$ 2.000,00 por mês mais
uma comissão de R$ 500,00 para cada carro vendido e mais 1% do valor da venda. Elabore um
algoritmo para calcular e imprimir o salário do vendedor num dado mês recebendo como dados
de entrada o nome do vendedor, o número de carros vendidos e o valor total das vendas deste
vendedor.

```python
print("Vamos calcular seu salario de acordo com a quantidade de vendas:")

nome = input("Qual seu nome?")
venda = float(input("Quantos carros você vendeu? "))
salario = 2000

print("Ola ", nome, ". Seu salário de acordo com a quantidade de carros vendidos esse mês é: ", ( ((salario + (venda * 500) ) * 0.10) + (venda * 500)) + salario  )
```

#8º Exercicio

Calcule a média de um aluno na disciplina de ADM. Para isso, solicite o nome do aluno, a
nota da prova e a nota do trabalho. Sabe-se que a nota da prova tem peso 2 e a nota do
trabalho tem peso 1. Mostre a média e escreva se ele foi aprovado ou reprovado.

```python
print("Vamos calcular a nota do aluno para saber se ele passou ou não:")

nome = input("Qual o nome do aluno?")

prova = float(input("Qual foi a nota da prova? "))
trabalho = float(input("Qual foi a  nota do trabalho? "))

media = ((prova*2) + trabalho )/3

if media > 6.5:
  print("O aluno ", nome, " passou! Sua média foi de:", media)
else:
  print("O aluno ", nome, " não passou! Sua média foi de:", media)
```
