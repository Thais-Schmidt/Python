##Uso da estrutura **if aninhada**

Quando precisamos resolver problemas mais complexos surge a necessidade de sninhar varios **ifs**, ou seja, utilizar **um if dentro do outro.**

#Exercicio 1

###A operadora de telefonia tchau oferece um plano de tarifas regressivas de acordo com o consumo em minutos.

###Para um consumo abaixo de 200 minutos, a tarifa é de R$0,20/min.

###De 200 a 400 minutos, a tarifa é de R$0,50/min.

###Acima de 400 minutos a tarifa é de R$0,10/min.

###Fazer um programa que pergunte a quantidade de minutos utilizada e calcule o valor a ser pago.

```python
minutos = int(input('Digite a quantidade de minutos: '))
if minutos < 200:
  tarifa = 0.20 #executa se (MINUTOS < 200 ) for True
else: #executa se (MINUTOS < 200 ) for False
  if minutos < 400: #If aninhado no item anterior
    tarifa = 0.15
  else:
    tarifa = 0.10

valor = minutos * tarifa
print('O valor a ser pago é de R$ ', valor)
```

#Exercicio 3

###Crie um programa em Python que solicite que o usuario informe um numero entre 1 a 7 e exiba na tela:

###Segunda-feira caso o numero digitado seja 1;

###Terça-feira caso o numero digitado seja 2;

### e seguindo a mesma logica até Domingo

###Obs: Caso o numero que o usuario informou esteja fora do limite estabelecido, deve ser exibida a mensagem na tela indicando que o numero é invalido.



```python
dia = int(input('Digite um numero de 1 a 7 para saber o dia da semana: '))

if dia > 0:
  if dia > 7:
    print('Digite um dia valido!')
  else:
    if dia == 1:
      print('Segunda-feira')
    else:
      if dia == 2:
        print('Terça-feira')
      else:
        if dia == 3:
          print('Quarta-feira')
        else:
          if dia == 4:
            print('Quinta-feira')
          else:
            if dia == 5:
              print('Sexta-feira')
            else:
              if dia == 6:
               print('Sábado')
              else:
                if dia == 7:
                  print('Domingo')
else:
  print('Digite um dia valido!')
```

#Elif

### Uma alternativa ao uso da estrutura **if aninhada** é a estrutura **if elif else**

### O **elif** é simplesmente a contração de **else + if** e é usado para fazer a estrutura ficar mais clara

### Vamos agora alterar os programas feitos para usar esta estrutura **if elif else**

#Exercicio 4 (ex_1)

```python
minutos = int(input('Digite a quantidade de minutos: '))
if minutos < 200:
  tarifa = 0.20 #executa se (MINUTOS < 200 ) for True
elif minutos < 400: #If aninhado no item anterior
  tarifa = 0.15
else:
  tarifa = 0.10
valor = minutos * tarifa
print('O valor a ser pago é de R$ ', valor)
```

### A partir da versão 3.10 do Python, uma alternativa mais elegante ao uso da estrutura **if** aninhada e da estrutura **if elif else** foi adicionada, trata-se do **$match$ - $case$**

### Podemos usar esta estrutura $match$ - $case$ para escolher o código que será executado em função dos valores. Vejamos o exemplo abaixo:

#Exemplo:

```python
dia = int(input('Digite um numero de 1 a 7 para saber o dia da semana: '))

match dia :
  case 1:
    print('Segunda-feira')
  case 2:
    print('Terça-feira')
  case 3:
    print('Quarta-feira')
  case 4:
    print('Quinta-feira')
  case 5:
    print('Sexta-feira')
  case 6:
    print('Sábado')
  case 7:
    print('Domingo')
  case _:
    print('Digite um dia valido!')
```
