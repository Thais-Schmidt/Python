#Operadores Relacionais

Os operadores relacionais são usados para comparar valores.

O resultado poder ser:

**True** (verdadeiro ou 1) se a condição for satisfeita;

**False** (falso ou 0) se a condição não for atendida.

```python
num1 = 5;
num2 = 9;
```

```python
print('O valor de num1 é ', num1, '\nO valor de num2 é ', num2)
```

```python
#testando se o valor de num1 é igual a 5
if num1 == 5:
  print("O valor guardado na variavel é igual a 5.")
```

```python
#testando se o valor de num2 é igual a 5
if num2 == 5:
  print ("O valor guardado na variavel é igual a 5.")
else:
  print ("O valor guardado na variavel \033[1mNÃO\033[0m é igual a 5.")
```

```python
num1 == num2
```

```python
num1 != num2
```

```python
#Colocar texto em negrito
print("\033[1mTexto em negrito\033[0m")
```

#Testando operador DIFERENTE DE

---



```python
if num1 != 5:
  print("O valor guardado na variavel é \033[1mdiferente\033[0m de 5.")
```

```python
if num2 != 5:
  print("O valor guardado na variavel é \033[1mdiferente\033[0m de 5.")
```

```python
if num1 != num2:
  print ("O valor da variavel num1 é \033[1mdiferente\033[0m da variavel num2")
else:
  print ("O valor da variavel num1 é \033[1migual\033[0m a variavel num2")
```

#Testando operador MAIOR QUE



```python
if num1 > 5:
  print("O valor guardado na variavel é \033[1MAIOR\033[0m de 5.")
```

```python
if num2 > 5:
  print("O valor guardado na variavel é \033[1mMAIOR\033[0m de 5.")
```

```python
if num1 > num2:
  print ("O valor da variavel num1 é \033[1mÉ MAIOR QUE\033[0m a variavel num2")
else:
  print ("O valor da variavel num1 \033[1mNÃO é MAIOR QUE\033[0m a variavel num2")
```

#Testando operador MENOR QUE

```python
if num1 < 5:
  print("O valor guardado na variavel é \033[1mMENOR\033[0m que 5.")
```

```python
if num2 < 5:
  print("O valor guardado na variavel é \033[1mMENOR\033[0m que 5.")
```

```python
if num1 < num2:
  print ("O valor da variavel num1 é \033[1m MENOR QUE\033[0m a variavel num2")
else:
  print ("O valor da variavel num1 é \033[1m MAIOR QUE\033[0m a variavel num2")
```

#Testando operador MAIOR OU IGUAL

```python
if num1 >= 5:
  print("O valor guardado na variavel é \033[1mMAIOR OU IGUAL\033[0m que 5.")
```

```python
if num2 >= 5:
  print("O valor guardado na variavel é \033[1mMAIOR OU IGUAL\033[0m que 5.")
```

```python
if num1 >= num2:
  print ("O valor da variavel num1 é \033[1mMAIOR OU IGUAL\033[0m a variavel num2")
else:
  print ("O valor da variavel num1 é \033[1mMENOR\033[0m que a variavel num2")
```
