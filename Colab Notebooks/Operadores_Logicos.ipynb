Os operadores lógicos são usados para executar testes lógicos ou avaliar expressoes booleanas.

## Lógica E

Utiliza-se para isso o operador **and**

```python
num1 = 5
num2 = 9
```

```python
num1 > 2
```

```python
num1 > 8
```

```python
if num1>2 and num1<8:
  print("num1 é maior que 2 e menor que 8")
else:
  print("uma das condições é falsa!")
```

A expressão testada: **num1>2 and num1<8**, somente será verdadeira se as duas comparações forem verdadeiras;

```python
print(num1>2)
```

```python
teste = num1<8
print(teste)
```

```python
teste = num1<8 and num1<8
print(teste)
```

#Lógica OU(or)

```python
#Um exemplo de utilização do operador OU é para testar se um numero é igual a um de dois valores aceitos
if num1==5 or num1==9:
  print("O valor da variavel num1 é 5 ou 9")
else:
  print("O valor da num1 não esta nas opções aceitaveis")
```

```python
if num2==2 or num2==9:
  print("O valor da variavel num2 é 5 ou 9")
else:
  print("O valor da num2 não esta nas opções aceitaveis")
```

```python
if num2==2 or num2==7:
  print("O valor da variavel num2 é 5 ou 9")
else:
  print("O valor da num2 não esta nas opções aceitaveis")
```

#Logica NAO

Utiliza-se para isso o operador **not**

Sua função é inverter o resultado lógico de **True** para **False** ou de **False** para **True**



```python
not(True)
```

```python
not(False)
```

```python
if teste:
  pass
else:
  print("O valor de num2 esta na faixa de 2 a 8")
```

```python
if not(teste):
  print("O valor de num2 esta na faixa de 2 a 8")
```

```python
Aprovado = True
if Aprovado:
  print("Parabens")
```

```python
Aprovado = False
if not(Aprovado):
  print("Reprovado")
```
