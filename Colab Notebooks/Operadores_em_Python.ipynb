##Conversão de tempo

Estes operadores são uteis para se fazer a conversão de tempo. Por exemplo, se eu precisar dividir um turno de 9 horas em 4 partes iguais, qual o periodo de tempo que cada parte terá?

```python
print("Cada parte terá:", (9/4), "horas.")
```

Mas esse texto não esta muito claro,seria melhor informar o periodo em horas e minutos.

```python
print("Cada parte terá:", 9//4, "horas.")
```

Precisamos agora converter o resto da divisão que estará em horas fracionárias, para minutos.

```python
print("Apos dividir as 9h em 4 pedaços de 2h, teremos o resto: ", (9%4), "hora.")
```

```python
print("Apos dividir as 9h em 4 pedaços de 2h, teremos o resto: ", (9%4)*60, "minutos.")
```

```python
print("Cada parte terá:", 9//4, "horas e ", int((9%4)*60/4), "minutos." )
```

#Exponenciação

A potenciação (ou exponenciação) é uma operaçõa onde um dado numero é multiplicado por ele mesno uma quantidade n de vezes.

```python
5**2
```

```python
#Calculando 5 elevado ao quadrado:
print("5² = 5*5 = ", 5**2)
```

```python
#Calculando 2 elevado ao cubo:
print("2³ = 2*2*2 = ", 2**3)
```
