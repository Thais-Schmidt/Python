

```
1. Escreva um programa que pergunte um número inteiro ao usuário e calcule o fatorial deste número.
O fatorial de um número é o produto de todos os seus antecessores, por exemplo:
fatorial de 5 = 5! = 5*4*3*2*1 = 120
```



````python
def calcular_fatorial(numero):
    resultado = 1
    for i in range(1, numero + 1):
        resultado *= i
    return resultado

numero = int(input("Digite um número: "))
resultado = calcular_fatorial(numero)
print("O fatorial de", numero, "é", resultado)
````
